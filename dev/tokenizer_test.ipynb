```python
from transformers import CLIPTokenizer

tokenizer = CLIPTokenizer.from_pretrained("openai/clip-vit-large-patch14")
```

```python
prompt = "Hello, World!"

tokens = tokenizer(
    prompt,
    padding="max_length",
    max_length=77,
    truncation=True,
    return_tensors="pt"
)

print(tokens["input_ids"])
```

```python
from sd15.model.clip import CLIP

clip = CLIP()
```

```python
from sd15.convert_weights import load_clip
import os

ckpt_path = os.path.join("..", "data", "weights", "v1-5-pruned-emaonly.safetensors")

load_clip(ckpt_path, clip)
```
